# Imports

```python
import sys

from matplotlib import pyplot as plt

sys.path.append("../../src")
import datasets
```

# Load Dataset

```python
bit_mix = datasets.bit_mix
training_df, is_testing_df, oos_testing_df = bit_mix.load_data(verbose=False)
```

# Feature Distribution

```python
for name,df in {
    "Training Data":training_df,
    "Testing Data (In Sample)":is_testing_df,
    "Testing Data (Out of Sample)":oos_testing_df
}.items():
    fig,axes = datasets.visualizations.plot_feature_grid(
        df,
        df.columns,
        target=df.columns[-1],
    )
    fig.suptitle(name)
    plt.tight_layout()
    plt.show()
```

# Features vs. Target

```python
for name,df in {
    "Training Data":training_df,
    "Testing Data (In Sample)":is_testing_df,
    "Testing Data (Out of Sample)":oos_testing_df
}.items():
    fig,axes = datasets.visualizations.plot_features_vs_target(
        df,
        df.columns[:-1],
        target=df.columns[-1],
    )
    fig.suptitle(name)
    plt.tight_layout()
    plt.show()
```

# Pairwise Feature Plots

```python
for name,df in {
    "Training Data":training_df,
    "Testing Data (In Sample)":is_testing_df,
    "Testing Data (Out of Sample)":oos_testing_df
}.items():
    fig,axes = datasets.visualizations.plot_feature_pairs(
        df,
        df.columns,
    )
    fig.suptitle(name)
    plt.tight_layout()
    plt.show()
```

# Distribution Comparison

## `training_df` vs `is_testing_df`

```python
fig,axes,results = datasets.visualizations.compare_feature_distributions_grid(
    training_df,
    is_testing_df,
    training_df.columns
)
fig.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```

## `training_df` vs `oos_testing_df`

```python
fig,axes,results = datasets.visualizations.compare_feature_distributions_grid(
    training_df,
    oos_testing_df,
    training_df.columns
)
fig.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
